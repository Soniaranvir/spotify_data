```python
# Initialize lists to hold album and artist data
album_list = []
artist_list = []

# Loop through the tracks to extract album and artist information
for row in data['items']:
    # Extract album details
    album_id = row['track']['album']['id']
    album_name = row['track']['album']['name']
    album_release_date = row['track']['album']['release_date']
    album_url = row['track']['album']['external_urls']['spotify']
    album_artists = [artist['name'] for artist in row['track']['album']['artists']]
    
    # Add album data to album_list
    album_element = {
        'album_id': album_id,
        'album_name': album_name,
        'release_date': album_release_date,
        'album_url': album_url,
        'album_artists': ", ".join(album_artists)
    }
    album_list.append(album_element)
    
    # Extract artist details (if not already included)
    for artist in row['track']['album']['artists']:
        artist_id = artist['id']
        artist_name = artist['name']
        artist_element = {
            'artist_id': artist_id,
            'artist_name': artist_name
        }
        artist_list.append(artist_element)

# Print to verify data extraction
print("Album and Artist Data Extraction Complete")
```

```python
# Create DataFrames for albums and artists
album_df = pd.DataFrame.from_dict(album_list)
artist_df = pd.DataFrame.from_dict(artist_list)

# Display album and artist DataFrames
print("Album DataFrame:")
print(album_df.head())
print(album_df.info())

print("\nArtist DataFrame:")
print(artist_df.head())
print(artist_df.info())
```

```python
# Optional: Drop duplicates from both DataFrames
album_df = album_df.drop_duplicates(subset=['album_id'])
artist_df = artist_df.drop_duplicates(subset=['artist_id'])

# Display the cleaned DataFrames
print("\nCleaned Album DataFrame:")
print(album_df.head())
print("\nCleaned Artist DataFrame:")
print(artist_df.head())
```
